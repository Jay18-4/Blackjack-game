```python
import random
from IPython.display import clear_output
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True
```

```python
class Card:
    def __init__(self,suits,ranks):
        self.suits = suits
        self.ranks = ranks
        self.values = values[ranks]

    def __str__(self):
        return f'{self.ranks} of {self.suits}'
```

```python
class Deck:
    def __init__(self):
        self.all_cards = []
        for suit in suits:
             for rank in ranks:
                str(self.all_cards.append(Card(suit,rank)))

    def __str__(self):
        return f'all {self.all_cards}'
                                      
    def shuffel(self):
        random.shuffle(self.all_cards)

    def deal_one(self):
        return (str(self.all_cards.pop()))
```

```python
class Hand:
    def __init__(self):
        self.cards = []  
        self.value = 0   
        self.aces = 0    
        
    def add_card(self,card):
        self.cards.append(card)
        rank = card.split(' ')[0]
        self.value += values[rank]
        if rank == 'Ace':
            self.aces += 1
        self.adjust_for_ace()
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces > 0:
            self.value -= 10
            self.aces -= 1

    def get_cards(self):
        return self.cards

    def get_value(self):
        return self.value

    def __str__(self):
        return f'Hand:{self.cards} (Value:{self.value})'
```

```python
class Chips:
    def __init__ (self,chips):
        self.chips = chips
        self.bet = 0

    def win_bet(self):
        self.chips += self.bet
        # print(f' YOU WON THE BET! Chips remaining:{self.chips}')

    def lose_bet(self):
        self.chips -= self.bet
        # print(f' YOU LOST THE BET! Chips remaining:{self.chips}')
        
    def take_bet(self):
        while True:
            try:
                bet_amount = int(input('Place Bet'))
                if 0 < bet_amount <= self.chips:
                    self.bet = bet_amount
                    break
                elif bet_amount == 0:
                    self.bet = bet_amount
                    print("Invalid Chips Amount!")
                    break
                else:
                    print(f'Invalid amount pls pick between 1 and {self.chips}')
            except ValueError:
                print("invalid Input Please Select A Whole Number") 
                
    def __str__(self):
        return f"Player's Chips {self.chips}"
```

```python
def player_hit_stand(deck,hand):
    global playing
    while playing:
        player_hit_or_stand = input("TAKE A HIT! Y/N ").upper()
        if player_hit_or_stand == 'Y':
            clear_output()
            hand.add_card(deck.deal_one())
            print(f"Player Hand {show_some(player_hand,dealer_hand)}")
            if hand.get_value() > 21:
                return False
                playing = False
        elif player_hit_or_stand == 'N':
            playing = False
            break
```

```python
def dealer_hit(deck,hand):
    while hand.get_value() < 17:
        hand.add_card(deck.deal_one())
        
```

```python
def show_some(player_hand,dealer_hand):
    player_hand = [player_hand.get_cards(), f'Value: {player_hand.value}']
    dealer_hand = ['Hidden'] + [dealer_hand.get_cards()[1]]
    return f'Player:{player_hand}\nDealer:{dealer_hand}'

def show_all(player_hand,dealer_hand):
    player_hand = [player_hand.get_cards(), f'(Value: {player_hand.value})']
    dealer_hand = [dealer_hand.get_cards(), f'(Value:{dealer_hand.value})']
    return f'Player:{player_hand}\nDealer:{dealer_hand}'
    
```

```python
def player_bust():
    return player_hand.get_value() > 21

def dealer_bust():
    return dealer_hand.get_value() > 21

def player_wins():
    return player_hand.get_value() > dealer_hand.get_value() and player_hand.get_value() <= 21 and dealer_hand.get_value() <= 21
    
def dealer_wins():
    return player_hand.get_value() < dealer_hand.get_value() and player_hand.get_value() <= 21 and dealer_hand.get_value() <= 21
            
def push():
    return player_hand.get_value() == dealer_hand.get_value() and player_hand.get_value() <= 21 and dealer_hand.get_value() <= 21
```

```python
player_chips = Chips(100)
```

```python
while True:
    # Print an opening statement
    print("Welcome To Black_Jack")
    
    # Create & shuffle the deck, deal two cards to each player
    deck = Deck()
    deck.shuffel()
    player_hand = Hand()
    dealer_hand = Hand()
    for x in range(2):
        player_hand.add_card(deck.deal_one())
        dealer_hand.add_card(deck.deal_one())
        
    # Set up the Player's chips
    player_chips.chips
    
    # Prompt the Player for their bet
    print(player_chips)
    player_chips.take_bet()
    
    # Show cards (but keep one dealer card hidden)
    print(show_some(player_hand,dealer_hand))

    playing = True
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        player_turn = player_hit_stand(deck,player_hand)
         
        # Show cards (but keep one dealer card hidden)
        # print(show_some(player_hand,dealer_hand))
         
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_turn == False :
            player_bust()
            player_chips.lose_bet()
            print("Player BUST!")
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_bust() == False:
        dealer_turn = dealer_hit(deck,dealer_hand)
    
        # Show all cards
        print(show_all(player_hand,dealer_hand))
        # Run different winning scenarios
        if player_wins():
            player_chips.win_bet()
            print("Congrats Player WON!")
        if dealer_wins():
            player_chips.lose_bet()
            print("Player LOST!")
        if player_bust():
            player_chips.lose_bet()
            print("Player BUST!")
        if dealer_bust():
            player_chips.win_bet()
            print("Dealer BUST!. Congrats Player WON!")
        if push():
            print("TIE!")
                
    
    # Inform Player of their chips total 
    print(player_chips)
    # Ask to play again
    play_again = input('Would Like To Play Again? Y/N').upper()
    if play_again == 'Y':
        clear_output()
        playing = True
    else:
        break
       
```
